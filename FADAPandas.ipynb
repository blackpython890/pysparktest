### Question 1 Solution

```python
orders_store = pd.merge(orders, store, left_on="store_id", right_on="id")
```

```python
orders_store.head()
```

```python
orders_store['order_date'] = pd.to_datetime(orders_store['order_date'])
```

```python
orders_store['years'] = orders_store['order_date'].dt.year
```

```python
orders_store['month']=orders_store['order_date'].dt.month
```

```python
orders_store.head()
```

```python
k = orders_store.groupby(['years','month','name'])\
                .agg({'store_id':"count",'total':sum})\
                .reset_index()
```

```python
k.columns = ['years','month','name','TotalOrders','TotalRevenue']
```

```python
k.head(20)
```

```python
k.to_csv('Quest1Soln.csv')
```

### Question 2 Solution

```python
cust_order = pd.merge(customer, orders, left_on="id", right_on="customer_id")
```

```python
s = cust_order.groupby(['first_name','last_name','email','customer_id'])\
    .customer_id.count()\
    .reset_index(name='OrderPlacedByUser')
```

```python
s.loc[s['OrderPlacedByUser']<10,['first_name','last_name','email','OrderPlacedByUser']]
```

```python
s.loc[s['OrderPlacedByUser']<10,['first_name','last_name','email','OrderPlacedByUser']]\
  .to_csv('Quest2Soultion.csv')
```

### Question3 Solution

```python
import hashlib
```

```python
df3 = s.loc[s['OrderPlacedByUser']<10,['first_name','last_name','email','OrderPlacedByUser']]
```

```python
df3['email'] = [hashlib.md5(val.encode('UTF-8')).hexdigest() for val in df3['email']]
```
